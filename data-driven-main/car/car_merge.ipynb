```python
# 完整的代码，包括加载数据和填充参数的过程

import pandas as pd

# 加载 car.xlsx 文件
car_data_path = 'car.xlsx'
car_data = pd.read_excel(car_data_path)

# 加载 most-popular-car-by-state.xlsx 文件
popular_car_data_path = 'most-popular-car-by-state.xlsx'
popular_car_data = pd.read_excel(popular_car_data_path)

# 调整 popular_car_data 的列名以便于自动化处理
popular_car_data.columns = ['Car-Name', 'Number_of_seats', 'Starting_MSRP', 'Safety_score', 'Fuel']

# 定义一个函数来为给定的车型名填充参数
def fill_car_parameters(row):
    for i in range(1, 6):  # 对于每种车型
        car_name_col = f'Car{i}-Name'
        car_name = row[car_name_col]
        if pd.notna(car_name):
            # 查找与当前车型名匹配的行
            match = popular_car_data[popular_car_data['Car-Name'].str.lower() == car_name.lower()]
            if not match.empty:
                # 填充四个参数
                row[f'Car{i}-Number of seats'] = match['Number_of_seats'].values[0]
                row[f'Car{i}-Starting MSRP'] = match['Starting_MSRP'].values[0]
                row[f'Car{i}-Safety score'] = match['Safety_score'].values[0]
                row[f'Car{i}-fuel'] = match['Fuel'].values[0]
    return row

# 应用填充函数到每行
car_data_filled = car_data.apply(fill_car_parameters, axis=1)

# 显示填充后数据的前几行以验证结果
car_data_filled.head()

# 保存填充后的数据到新的 Excel 文件
car_data_filled.to_excel('car_filled.xlsx', index=False)
```
