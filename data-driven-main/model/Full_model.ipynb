```python
import pandas as pd
test=pd.read_excel('test.xlsx')
test_fuel = test.drop(['State','Average Number of Seats','Average Starting MSRP','Average Safety Score'], axis=1)
test_seats = test.drop(['State','Most Common Fuel Type','Average Starting MSRP','Average Safety Score'], axis=1)
test_price = test.drop(['State','Most Common Fuel Type','Average Number of Seats','Average Safety Score'], axis=1)
test_safety = test.drop(['State','Most Common Fuel Type','Average Number of Seats','Average Starting MSRP'], axis=1)
```
